# Martketing Customer Analysis - Info

```python

import pandas as pd
import numpy as np


import warnings 
warnings.filterwarnings('ignore')


import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns

customer_mkt = pd.read_csv(r"C:\Users\haiss\Downloads\lab-customer-analysis-round-2\files_for_lab\csv_files\marketing_customer_analysis.csv")
customer_mkt
```

```python
cols = []
for i in range(len(customer_mkt.columns)):
    cols.append(customer_mkt.columns[i].lower().replace(' ', '_')) 
customer_mkt.columns = cols
customer_mkt
```

## Total number of response

```python
sns.histplot(customer_mkt['response'])
plt.show()
```

## Response rate by the sales channel

```python
customer_mkt=customer_mkt.groupby(['response','sales_channel']).size()
customer_mkt=customer_mkt.unstack()
customer_mkt.plot(kind='bar')
```

# Response rate by the total claim amount.

```python
sns.barplot(x=customer_mkt['response'], y=customer_mkt['total_claim_amount'], data=customer_mkt, estimator='mean') 
plt.show()
```

## Response rate by income.

```python
customer_mkt=customer_mkt.groupby(['response','income']).size()
customer_mkt.plot(kind='bar')
```
